Musimy go wczytać, jako macierz wektorów 4-wymiarowych (RGB i kanał przezroczystości), a następnie będziemy zliczać występowanie każdego z wektorów. To zliczenie posłuży nam do znalezienia rozkładu prawdopodobieństwa:

```python
import numpy as np
from PIL import Image
```

```python
im_frame = Image.open('../images/checkerboard.png')
np_frame = np.array(im_frame.getdata())
```

Stwórzmy funkcję, która z wektora `numpy` zwraca występujące elementy i ich prawdopodobieństwa wystąpienia:

```python
def rozklad(array):
    elementy, zliczenia = np.unique(array, return_counts=True, axis=0)
    prawdopodobienstwa = np.divide(zliczenia, array.shape[0])
    return (elementy, prawdopodobienstwa)
    
```

```python
elementy, prawdopodobienstwa = rozklad(np_frame)
elementy_indeksy = []

for i in range(len(elementy)):
    elementy_indeksy.append(str(i))

dictionary = {}
for A, B in zip(elementy_indeksy, prawdopodobienstwa):
    dictionary[A] = B
dictionary

dit.Distribution(dictionary)
```

Otrzymaliśmy rozkład prawdopodobieństwa wystapienia danego koloru w podanym obrazku. Właśnie w ten sposób działa ta biblioteka - tworzymy rozkłady, a następnie dla nich można wyliczać własności z teorii informacji.
