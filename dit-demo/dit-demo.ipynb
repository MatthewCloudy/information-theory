Na początek zainstalujmy bibliotekę za pomocą komendy `pip install dit` wpisanej w konsoli.


```python
import dit
```

Stwórzmy rozkład przykładowego źródła - kostki sześciennej:

```python
die_throws = {'1': 1/6, '2': 1/6, '3': 1/6, '4': 1/6, '5': 1/6, '6': 1/6}
die_distribution = dit.Distribution(die_throws)
```

Możemy wypisać własności rozkładu:

```python
die_distribution
```

Policzmy teraz entropię dla tego rozkładu:

```python
dit.shannon.entropy(die_distribution)
```
